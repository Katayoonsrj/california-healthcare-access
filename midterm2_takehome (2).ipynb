```python
# Fix data types
df
```

```python
df['uninsured_rate'] = df['uninsured_rate'].str.replace('%', '')
df['uninsured_rate'] = df['uninsured_rate'].astype(float)
df['uninsured_rate'] = df['uninsured_rate']/100
df['unemployment_rate'] = df['unemployment_rate'].str.replace('%', '')
df['unemployment_rate'] = df['unemployment_rate'].astype(float)
df['unemployment_rate'] = df['unemployment_rate']/100
df['poverty_rate'] = df['poverty_rate'].str.replace('%', '')
df['poverty_rate'] = df['poverty_rate'].astype(float)
df['poverty_rate'] = df['poverty_rate']/100
df['median_income'] = df['median_income'].str.replace('\D', '', regex=True)
df['median_income'] = df['median_income'].astype(float)
df['Population'] = df['Population'].str.replace(',', '')
df['Population'] = df['Population'].astype(int)
df
```

### 7.1 Visualization 1

*What question does this plot answer?*

```python
# Visualization 1
yearly = df.groupby('Year')[['median_income']].mean().reset_index()

sns.lineplot(data=yearly, x='Year', y='median_income')
plt.title('Median Income Over Time')
plt.show()
```

**Interpretation:** *(2–3 sentences)*

*From this graph, we can see that our median income for all counties has steadily increased over the years. However, this does not mean that purchasing power has increased. This could be a sign of inflation instead of actual economic growth*

### 7.2 Visualization 2 *(must be faceted or grouped)*

*What subgroup comparison are you making, and why is it meaningful for your research question?*

```python
# Visualization 2 — faceted or grouped
# Hint: hue=, col=, row= in Seaborn; facet_wrap() or facet_grid() in plotnine
sns.lineplot(data=df, x='Year', y='cost_burden_rate', hue='poverty_level')
plt.title('Cost Burden Rate by Poverty Level Over Time')
plt.show()
```

**Interpretation:** *(2–3 sentences — what does the subgroup comparison reveal?)*

*The visualization shows that cost burden rates increase from the pre covid (around 2018–2019)to post covid years (2020–2023). This shows that housing affordability over time is becoming much worse. Although both low and high poverty counties show this increase, high poverty counties consistently maintain higher cost burden rates every year. The timing shows a positive correlation between post covid and the cost burden rising.*

### 7.3 Visualization 3

*What question does this plot answer?*

```python
# Visualization 3
yearly = df.groupby('Year')[['poverty_rate', 'uninsured_rate']].mean().reset_index()

sns.lineplot(data=yearly, x='Year', y='poverty_rate', label='Poverty Rate')
sns.lineplot(data=yearly, x='Year', y='uninsured_rate', label='Uninsured Rate')

plt.title('Poverty vs Uninsured Rates Over Time')
plt.ylabel('Rate')
plt.legend()
plt.show()
```

**Interpretation:** *(2–3 sentences)*

*We actually do see a decrease in the poverty and uninsured rate here, especially from 2018 to 2019 (pre covid). However, during 2020 on onward (post covid), the graph starts leveling off. This suggests that while income growth may initially reduce poverty, its impact weakens over time. Overall, increases in income do not lead to consistent or sustained decreases in poverty.*
